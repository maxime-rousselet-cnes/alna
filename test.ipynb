```python
from sympy import integrate, symbols, I, lambdify
from numpy import logspace, zeros_like
from mpmath import exp
from matplotlib.pyplot import semilogx

tau_0, tau_m, tau_M, alpha, omega, tau = symbols("tau_0 tau_m tau_M alpha omega tau")
test_integrand = (tau / tau_0) ** alpha / (1 + I * omega * tau) / tau
expr = integrate(test_integrand, (tau, tau_m, tau_M)).diff(alpha)
f = lambdify((tau_m, tau_M, alpha, tau_0, omega), expr, modules=[{"exp_polar": exp}, "mpmath"])
periods = logspace(1, 10, 30, 10)
omega_values = 1 / periods
f_values = zeros_like(a=periods, dtype=complex)

for omega_index, omega_value in enumerate(omega_values):

    f_values[omega_index] = f(3600, 86164 * 365, 0.26, 1, omega_value)

semilogx(periods, f_values.real)
```

```python
semilogx(periods, f_values.imag)
```
