```python
from sympy import symbols, I, integrate, lerchphi, hyper, I, exp_polar, Expr
from sympy import pi as sympy_pi
from IPython.display import display
from alna.rheological_formulas import find_tau_m_sup

alpha, omega, tau, tau_0, frequency_unit, tau_m_inf, delta, q_mu = symbols(
    r"\alpha \omega \tau \tau_0 f_{unit} \tau^{m-inf} \Delta q_\mu"
)
a = I * omega * frequency_unit
f = integrate(
    (tau / tau_0) ** alpha / (1 + a * tau) / tau,
    (
        tau,
        tau_m_inf,
        find_tau_m_sup(
            omega_m_inf=1 / tau_m_inf,
            period_unit=1 / frequency_unit,
            alpha=alpha,
            delta=delta,
            q_mu=q_mu,
        ),
    ),
)


def rewrite_alpha_integral(expression: Expr) -> Expr:
    """
    Needed to rewrite the automatically derived expresison from sympy.
    """

    return expression.xreplace(rule={exp_polar(3 * I * sympy_pi / 2): -I}).simplify()


def rewrite_lerchphi_to_hyper(expression: Expr) -> Expr:
    """
    Fast, structure-aware rewrite replacing lerchphi(z, s, a)
    with hypergeometric equivalents for s = 1 or 2.
    """

    def rewrite(expr: Expr):

        if isinstance(expr, lerchphi):

            z, s, a = expr.args

            if s == 0:

                return 1 / (1 - z)

            if s == 1:

                return (1 / a) * hyper([1, a], [a + 1], z)

            if s == 2:

                return (1 / a**2) * hyper([1, 1, a], [2, a + 1], z)

        return expr

    return expression.replace(lambda e: isinstance(e, lerchphi), rewrite)


display(f)
print()
display(rewrite_alpha_integral(expression=f))
print()
display(rewrite_alpha_integral(expression=f).diff(alpha))
print()
display(rewrite_lerchphi_to_hyper(expression=rewrite_alpha_integral(expression=f).diff(alpha)))
# display(rewrite_lerch(expr=f.diff(alpha), alpha=alpha))
```
